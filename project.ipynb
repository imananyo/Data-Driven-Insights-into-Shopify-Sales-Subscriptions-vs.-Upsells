```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("advanced_shopify_sales_data.csv")

# Preprocess
date_cols = ["Order Date", "Subscription Start Date", "Subscription End Date"]
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

bool_cols = ["Is Upsell", "Churned", "Customer Email Opt-in", "Cart Abandonment", "Is First Purchase",
             "Bundle Purchase", "Review Left", "Gift Order", "Affiliate Source", "Is Repeat Subscription"]
for col in bool_cols:
    df[col] = df[col].map({"Yes": True, "No": False})

# Set seaborn style
sns.set(style="whitegrid", palette="muted")
```

```python
# Revenue Breakdown by Product Type 
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="Product Type", y="Revenue", estimator=sum, ci=None)
plt.title("Total Revenue by Product Type")
plt.ylabel("Total Revenue ($)")
plt.xlabel("Product Type")
plt.tight_layout()
plt.show()
```

```python
# Churn vs. Active Subscription Count 

plt.figure(figsize=(6, 4))
sns.countplot(data=df[df["Subscription Status"].notna()], x="Subscription Status")
plt.title("Subscription Status Distribution")
plt.xlabel("Status")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

```

```python
# Revenue by Traffic Source 
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="Traffic Source", y="Revenue", estimator=sum, ci=None)
plt.title("Revenue by Traffic Source")
plt.ylabel("Revenue ($)")
plt.xlabel("Traffic Source")
plt.tight_layout()
plt.show()

```

```python
# Revenue by Device Used 

plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x="Device Used", y="Revenue")
plt.title("Revenue Distribution by Device")
plt.xlabel("Device")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()
```
